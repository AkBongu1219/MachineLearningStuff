```python
import pandas as pd
import numpy as np
import math as m
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn import linear_model
```

```python
#Gamma and Regularization: If high, the classification boundary is very tight, if low, vice versa
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('L vs W')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color = 'r', marker = '+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'g', marker = 'o')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color = 'b', marker = 'x')
plt.show()
```

```python
x = df.drop(['target', 'flower_name'], axis = 'columns')
y = df.target
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.svm import SVC
svc = SVC()
```

```python
svc.fit(xtrain,ytrain)
```

```python
svc.score(xtrain,ytrain)
```

```python
svc.score(xtest,ytest)
```
