```python
df = pd.read_csv("C:\\Users\\Guest_User\\Downloads\\spam.csv")
df.head()
```

```python
df.groupby('Category').describe()
```

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x == 'spam' else 0)
df.head()
```

```python
df.drop(['Category'],axis = 'columns')
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(df.Message, df.spam, test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()
X_train_count = v.fit_transform(xtrain.values)
X_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,ytrain)
```

```python
emails = [
    'Hey jamaal, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(xtest)
model.score(X_test_count, ytest)
```

```python
wine = load_wine()
```

```python
df = pd.DataFrame(wine.data,columns=wine.feature_names)
df.head()
```

```python
df['target'] = wine.target
df
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(wine.data, wine.target, test_size=0.25, random_state=100)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```
