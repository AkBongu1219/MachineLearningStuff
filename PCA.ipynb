```python
import pandas as pd
import numpy as np
import math as m
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import svm, datasets
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
df = pd.DataFrame(digits.data, columns = digits.feature_names)
df.head()
```

```python
x = df
y = digits.target
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xs = scaler.fit_transform(x)
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(xs,y,test_size = 0.25, random_state = 2) 
mod = linear_model.LogisticRegression()
mod.fit(xtrain,ytrain)
mod.score(xtest,ytest)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
x_pca = pca.fit_transform(x)
x_pca.shape
```

```python
xtrain_pca,xtest_pca,ytrain_pca,ytest_pca = train_test_split(x_pca,y,test_size = 0.25, random_state = 2) 
mod_pca = linear_model.LogisticRegression(max_iter = 1000)
mod_pca.fit(xtrain_pca, ytrain_pca)
mod_pca.score(xtest_pca, ytest_pca)
```

```python
df[df.Cholesterol>(df.Cholesterol.mean()+3*df.Cholesterol.std())]#Used to eliminate outliers(Z-test apprently)
```

```python
df1 = df[df.Cholesterol<=(df.Cholesterol.mean()+3*df.Cholesterol.std())]
df1.shape
```

```python
df1.head()
```

```python
df2 = df1[df1.Oldpeak<=(df1.Oldpeak.mean()+3*df1.Oldpeak.std())]
df2.shape
df3 = df2[df2.RestingBP<=(df2.RestingBP.mean()+3*df2.RestingBP.std())]
df3.shape
```

```python
df4 = df3.copy()
df4.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df4.head()
```

```python
df5 = pd.get_dummies(df4, drop_first = True)
df5.head()
```

```python
x = df5.drop(['HeartDisease'], axis = 'columns')
y = df5['HeartDisease']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xs = scaler.fit_transform(x)
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(xs,y,test_size = 0.25, random_state = 2) 
mod = linear_model.LogisticRegression()
mod.fit(xtrain,ytrain)
mod.score(xtest,ytest)
```

```python
mod_svm = svm.SVC()
mod_svm.fit(xtrain,ytrain)
mod_svm.score(xtest,ytest)
```

```python
mod_rfc = RandomForestClassifier()
mod_rfc.fit(xtrain,ytrain)
mod_rfc.score(xtest,ytest)
```

```python
#USING PCA
from sklearn.decomposition import PCA

pca = PCA(0.95)
x_pca = pca.fit_transform(x)
x_pca
```

```python
xtrain_pca,xtest_pca,ytrain_pca,ytest_pca = train_test_split(x_pca,y,test_size = 0.2, random_state = 30) 
```

```python
mod = linear_model.LogisticRegression()
mod.fit(xtrain_pca,ytrain_pca)
mod.score(xtest_pca,ytest_pca)
```

```python
mod_svm = svm.SVC()
mod_svm.fit(xtrain_pca,ytrain_pca)
mod_svm.score(xtest_pca,ytest_pca)
```

```python
mod_rfc = RandomForestClassifier()
mod_rfc.fit(xtrain_pca,ytrain_pca)
mod_rfc.score(xtest_pca,ytest_pca)
```
