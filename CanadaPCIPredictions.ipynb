```python
import pandas as pd
import numpy as np
import math as m
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("C:\\Users\\Guest_User\\Downloads\\canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("PCI")
plt.title("Canada PCI Progression")
plt.plot(df.year, df['per capita income (US$)'], color = "green", marker = "o")
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
reg.predict(np.array([[2020]]))
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("PCI")
plt.title("Canada PCI over Time")
plt.scatter(df.year, df['per capita income (US$)'], color = "green", marker = "H")
plt.plot(df.year, reg.predict(df[['year']]), color = "blue")
plt.show()
```
